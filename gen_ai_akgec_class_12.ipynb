Spam Detection using Naive Bayes
1. Spam=1
2. Not Spam=0
3. Win Monney now = Spam
4. Meeting Scheduled tomorrow-Not Spam

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
emails=[
    "win money now",
    "limited offer claim prize",
    "meeting scheduled tomorrow",
    "project discussion at office"
]
labels=[1,1,0,0]
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(emails)
print("vocabulary:")
print(vectorizer.get_feature_names_out())
```

```python
print("Vector Form")
print(X.toarray())
```

```python
model=MultinomialNB()
model.fit(X,labels)
test_email="Claim prize money"
test_vector=vectorizer.transform([test_email])
print("Test Email Vector")
print(test_vector.toarray())
```

```python
prediction=model.predict(test_vector)
print("Predictions:",prediction)
```

Build a NLP based system using tf-idf and svm

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
news=[
    "India won cricket world cup",
    "Virat Kohli scored a century",
    "New AI software launched",
    "google released a new android update"
]
labels=[
    "sport",
    "sport",
    "technology",
    "technology"
]
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(news)
print("Vocabulary:")
print(vectorizer.get_feature_names_out())
```

```python
print("TF-IDF")
print(X.toarray())
```

```python
model=LinearSVC()
model.fit(X,labels)
test_news=[
    "AI software update released"
]
test_vector=vectorizer.transform(test_news)
print("test vector")
print(test_vector.toarray())
```

```python
prediction=model.predict(test_vector)
print("predicted category",prediction[0])
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
resumes=[
    "I have experience in AI,ML, data science ",
    "I am skilled in java,spring boot, backend development, api etc"
    "Experienced in HR Recruitment, Communication etc."
]
job="Looking for individual with AI,data science"
def clean_text(text):
  text=text.lower()
  text=re.sub(r'[^a-zA-Z0-9\s]','',text)
  return text
cleaned_resume=[clean_text(resume) for resume in resumes]
```

```python
clean_id=clean_text(job)
doc=cleaned_resume+[clean_id]
vectorizer=TfidfVectorizer(stop_words="english")
tfidf_matrix=vectorizer.fit_transform(doc)
similarity_score=cosine_similarity(tfidf_matrix[-1],tfidf_matrix[:-1])
print("cleaned resume")
print(cleaned_resume)
```

```python
print(vectorizer.get_feature_names_out())
```

```python
best_resume_index=similarity_score.argmax()
print("best resume matching index", best_resume_index+1)
print("best resume", resumes[best_resume_index])
```
