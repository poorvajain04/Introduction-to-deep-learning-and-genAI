```python
text="this is a very good product"
stop_words=["this","is","a","very"]
words=text.lower().split()
filtered=[]
for word in words:
  if word not in stop_words:
    filtered.append(word)
filtered
```

```python
t="the product is good and delivery is very fast"
stopwords=["the","is","and","very"]
w=t.split()
filter=[]
for wrd in w:
    if wrd not in stopwords:
        filter.append(wrd)
filter
```

**Sentiment Analysis**

```python
review="the product is good"
p=["good","amazing","best"]
n=["bad","poor","pathetic"]
text=review.lower()
sentiment="neutral"
for word in p:
  if word in text:
    sentiment="positive"
for word in n:
  if word in text:
    sentiment="negative"
sentiment
```

```python
review=[
    "good product",
    "bad experience",
    "average product"
    ]
p=["good","best","nice"]
n=["bad","poor","average"]
for r in review:
  sentiment="neutral"
  t=r.lower()
  for word in p:
    if word in t:
      sentiment="positive"
  for word in n:
    if word in t:
      sentiment="negative"
  print("sentiment:",sentiment)
```

```python
from numpy import vectorize
text=[
    "I LOVE NLP",
    "NLP IS USEFUL"
]
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(text)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
text=[
    "good good product",
    "bad product"
]
vector=CountVectorizer()
X=vector.fit_transform(text)
print(vector.get_feature_names_out())
print(X.toarray())
```

```python
#TF and IDF
#If any word appears many times, it can have have TF but a low IDF
#TF=TERM FREQUENCY
#IDF=INVERSE DOCUMENT FREQUENCY
texts=[
    "good product",
    "bad product",
    "good quality"
]
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf=TfidfVectorizer()
X=Tfidf.fit_transform(texts)
print(Tfidf.get_feature_names_out())
print(X.toarray())
```

```python
from sklearn.naive_bayes import MultinomialNB
messages=[
    "win money now",
    "limits offer wins",
    "hello friend",
    "how are you"
]
label=["spam","spam","not spam","not spam"]
vector=CountVectorizer()
X=vector.fit_transform(messages)
model=MultinomialNB()
model.fit(X,label)
test=["win offer"]
test_vector=vector.transform(test)
print(model.predict(test_vector[0]))
```
