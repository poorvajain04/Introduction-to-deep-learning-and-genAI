**Prompt Debugging Report**

```python
prompt='''
You are a business analyst.
Create a customer churn report for telecom company.
Use simple language.
give output in 5 bullet points.
include a real world example.
keep answer upto 150 words.
'''
score=0
feedback=[]
```

```python
if"i am" in prompt.lower():
  score+=1
else:
  feedback.append("Missing role")
if "create" in prompt.lower() or "explain" in prompt.lower() or "write" in prompt.lower():
  score+=1
else:
  feedback.append("Missing clear task")
if "customer churn" in prompt.lower() or "telecom" in prompt.lower():
  score+=1
else:
  feedback.append("Missing output format")
if "table" in prompt.lower() or "bullet" in prompt.lower():
  score+=1
if "under" in prompt.lower() or "simple language" in prompt.lower():
  score+=1
else:
  feedback.append("Missing constraint")
if "example" in prompt.lower():
  score+=1
else:
  feedback.append("Missing example requirement")
```

```python
print("prompt debugging report")
print("------------------------")
print("prompt score: ",score,"/6")
```

**Guardrails**

```python
user_query="Tell me how to hack an account"
blocked_words=["steal","hack","password","illegal"]
blocked=False
for word in blocked_words:
  if word in user_query:
    blocked=True
if blocked:
  print("Blocked:unsafe request")
else:
  print("allowed: safe request")
```

**Memory Simulation**

```python
memory={}
def chatbot(user_input):
  user_input_lower=user_input.lower()
  if "my name is" in user_input_lower:
    name=user_input.split("is")[-1].strip()
    memory["name"]=name
    return f"Nice to meet you, {name}"
  elif "what is my name" in user_input.lower():
    if "name" in memory:
      return f"Your name is: {memory['name']}"
    else:
      return "I dont know your name"
  else:
    return "Please tell me more"
print(chatbot("My name is Poorva"))
print(chatbot("What is my name?"))
```

```python
answer="Prompt engineering means writing clear instructions for an LLM"
expected_keywords=["prompt","instructions","llm"]
score=0
for word in expected_keywords:
  if word.lower() in answer.lower():
    score+=1
print("score: ",score,"/",len(expected_keywords))
```

**Output Format Validation**

```python
llm={
    "name": "Poorva",
    "age":20,
    "course":"llm"
}
required=["name","age","course"]
missing=[]
for data in required:
  if data not in llm:
    missing.append(data)
if len(missing)==0:
  print("Output format valid")
else:
  print("Missing fields are: ", missing)
```
