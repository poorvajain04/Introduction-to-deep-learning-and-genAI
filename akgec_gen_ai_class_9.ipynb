**BERT** **Implementation**





Convert given text to lower case and count the number of words.

```python
reviews=[
    "The Product works very well",
    "The Customer Service is Excellent",
    "Delivery was Delayed",
    "Packaging was Good"
]
for review in reviews:
  cleaned=review.lower()
  word_count=len(cleaned.split())
  print("Original:", review)
  print("lowercase:",cleaned)
  print("words:",word_count)
```

1. Count the words
2. Find the longest review
3. print review
4. print total words

```python
reviews=[
    "Movie was fantastic",
    "Movie was very interesting and enjoyable",
    "Not good",
    "Excellent acting and wonderful story"
]
sum=0
longest_review=""
for review in reviews:
  word_count=len(review.split())
  sum=sum+word_count
  print("review:",review)
  print("word_count:",word_count)
  if word_count>len(longest_review.split()):
    longest_review=review
print("longest review:",longest_review)
print("total words:",sum)


```

Sentiment Analysis

```python
reviews=[
    "This product is amazing",
    "Very poor quality",
    "Good packaging and fast delivery",
    "Worst Experience ever",
    "Average product"
]
positive=["good","amazing","fast"]
negative=["poor","worst","average"]
for review in reviews:
  text=review.lower()
  sentiment="Neutral"
  for word in positive:
    if word in text:
      sentiment="positive"
  for word in negative:
    if word in text:
      sentiment="negative"
  print(review)
  print("sentiment:",sentiment)
  print("-"*40) #print character 40 times
```
