```python
api_key=userdata.get("Gemini_api_key")
client=genai.Client(api_key=api_key)
print("Type exit to stop")
while True:
  user_input=input("You: ")
  if user_input.lower()=="exit":
    print("chat ended")
    break
  response=client.models.generate_content(
      model="gemini-2.5-flash",
      contents=user_input
  )
```

**Temperature:** Used to variate the response. Low no may result in not very good response (0-2)

```python
from google .colab import userdata
from google import genai
from google.genai import types
api_key=userdata.get("Gemini_api_key")
client=genai.Client(api_key=api_key)
response=client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Write a short note on cytokinins",
    config=types.GenerateContentConfig(
        temperature=2
    )
)
print(response.text)
```

**Streaming Response:** Model gives answer token by token. Improve user experience

```python
from google .colab import userdata
from google import genai
api_key=userdata.get("Gemini_api_key")
client=genai.Client(api_key=api_key)
stream=client.models.generate_content_stream(
    model="gemini-2.5-flash",
    contents="Explain the heuristic approach in machine learning with an example"
)
for chunk in stream:
  print(chunk.text,end="")
```
