# **Build a text classification pipeline for:**
 **1. Company receives thousands of customer reviews every minute, classify the reviews automatically**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
data={
    "review":[
        "I love this phone",
        "Worst Battery ever",
        "Amazing Camera quality",
        "Very slow performance",
        "Excellent display",
        "Waste of money"
    ],
    "label":[
        "positive",
        "negative",
        "positive",
        "negative",
        "positive",
        "negative"
    ]
}
df=pd.DataFrame(data)
model=Pipeline([
    ("tfidf",TfidfVectorizer()),
    ("classifier",LogisticRegression())
])
```

```python
model.fit(df["review"],df["label"])
```

```python
prediction=model.predict(["Good display quality"])
print(prediction)
```

```python
!pip install sumy
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
```

```python
text="""
Seeking to clarify the goals and usage implications of ethanol-blended petrol, the government has come out with a list of Frequently Asked Questions (FAQs). The move comes amid backlash and concerns about the effect of using E20 and other variants of ethanol-blended petrol on vehicles.
The government has said that the Ethanol Blended Petrol (EBP) Programme is a major component of India's energy strategy, designed to strengthen energy security, support the farming community and lower environmental impact by increasing the use of domestically produced renewable fuel. India reached the milestone of 20% ethanol blending in April . India runs over one lakh retail outlets through a vast netwo .
"""
```

```python
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
parser=PlaintextParser.from_string(text,Tokenizer("english"))
summarizer=LsaSummarizer()
summary=summarizer(parser.document,2)
for sentence in summary:
  print(sentence)
```
