```python
from google import genai
from google.colab import userdata
api_key = userdata.get("Gemini_api_key")
client = genai.Client(api_key=api_key)
product=[
    {
        "name":"Britania Treat",
        "price":45,
        "quantity":80,
        "in_stock": "yes"
    },
    {
        "name":"Sunfeast Bounce",
        "price":30,
        "quantity":45,
        "in_stock": "no"
    },
    {
        "name":"Hajmola Zeera",
        "price":20,
        "quantity":250,
        "in_stock": "no"
    },
    {
        "name":"Bingo Mad Angles",
        "price":50,
        "quantity":120,
        "in_stock": "yes"
    }
]
def product_agent(user_question):
  prompt=f"""
  You are a product information Engineer.
  You have access to the following product details
  {product}
  User Question: {user_question}
  Instructions:
1. Answer ONLY using the information in the product database.
2. Do NOT make up products, prices, quantities, or stock status.
3. If the answer cannot be found in the database, reply:
   "Sorry, this information is not available in the product database."
4. Present the answer in a clear, readable format.
5. If multiple products satisfy the query, list all of them.
6. Mention the product name, price, quantity, and stock status whenever relevant.
7. Be concise and professional.

Answer:
"""
  response=client.models.generate_content(
      model="gemini-2.5-flash",
      contents=prompt
)
  return response.text
question="Which eatable is currently in stock?"
answer=product_agent(question)
print(answer)
```

```python
products = {
    "Laptop": {
        "price": 60000,
        "stock": "yes",
        "discount": 15
    },
    "Phone": {
        "price": 50000,
        "stock": "yes",
        "discount": 10
    }
}

def get_price(product):
    return products[product]["price"]

def get_stock(product):
    return products[product]["stock"]

def get_discount(product):
    return products[product]["discount"]

def product_agent(question):
    question = question.lower()

    if "laptop" in question:
        product = "Laptop"
    elif "phone" in question:
        product = "Phone"
    else:
        return "Sorry, I couldn't identify the product."

    if "price" in question:
        price = get_price(product)
        return f"The price of the {product} is ₹{price}."

    elif "stock" in question or "available" in question:
        stock = get_stock(product)
        return f"The {product} is in stock: {stock}."

    elif "discount" in question:
        discount = get_discount(product)
        return f"The discount on the {product} is {discount}%."

    else:
        return "Sorry, I couldn't understand your question."

print(product_agent("What is the price of the laptop?"))
print(product_agent("Is the phone in stock?"))
print(product_agent("What is the discount on the laptop?"))
```
