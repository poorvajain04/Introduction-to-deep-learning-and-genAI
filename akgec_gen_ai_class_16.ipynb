# **A function that can generate AI prompt from product details**

```python
def product_prompt(product_name,target_customer,output_format):
    return f"""
    Act as a marketing expert.
    Create a promotional description for :
    Product:{product_name}
    Target Customer: {target_customer}
    Output Format: {output_format}
    Rules:
    Use persuasive but realistic language.
    Don't make fake promises.
    """
prompt= product_prompt(
    "Smart fitness watch",
    "Anyone aged between 25 to 40 years",
    "Headline + 3 benefits + CTA"
)
```

```python
print(prompt)
```

# **Create a prompt for E-mail (E-mail Generator)**

```python
def email_prompt(sender,receiver,subject_line):
  return f"""
  You are an email generator.
  Generate an email for:
  Sender:{sender}
  Receiver:{receiver}
  Subject:{subject_line}
  Rules:
  Keep the tone formal.
  Keep the email short and crisp
  """
```

```python
prompt=email_prompt("John","Anna","Increment in salary")
print(prompt)
```
