```python
!pip install tensorflow
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Dense,Input,LSTM
```

```python
X=np.array([
    [[1],[2],[1],[2],[1]],
    [[2],[1],[2],[1],[2]],
    [[5],[6],[5],[6],[5]],
    [[6],[5],[6],[5],[6]],
    [[1],[1],[2],[1],[1]],
    [[6],[6],[5],[6],[6]]
            ])
Y=np.array([0,0,1,1,0,1])
```

```python
X=X.reshape(6,5,1)
```

```python
model=Sequential()
```

```python
model.add(Input(shape=(5,1)))
model.add(SimpleRNN(8, activation="tanh"))
```

```python
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(X,Y,epochs=20)
model.summary()
```

```python
prediction = model.predict(
    np.array([[[5],[5],[6],[5],[6]]],dtype="float32")
)
prediction
```

Using LSTM model for a company to predict the purchases.


```python
X=np.array([
    [[100],[120],[150],[210],[130]],
    [[80],[81],[85],[90],[92]],
    [[150],[160],[190],[200],[210]],
    [[600],[500],[610],[510],[650]],
    [[700],[750],[720],[710],[780]],
    [[800],[900],[810],[840],[820]]
            ])
Y=np.array([0,0,1,1,0,1])
```

```python
model=Sequential()
model.add(LSTM(
    units=8,
    input_shape=(5,1)
))
```

```python
model.add(Dense(
    units=1,
    activation="sigmoid"
))
```

```python
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(X,Y,epochs=100)
```

```python
model.summary()
```

```python
test=np.array([[[500],[520],[640],[650],[780]]])
```

```python
test=test/900 # maximum value in data is 900 so used 900. Deep Learning use normalisation
```

```python
prediction=model.predict(test)
prediction
```

```python
if prediction[0][0]>0.6:
  print("previous member purchase")
else:
  print("no purchase")
```

Data set is small, sample size is small which is inefficient to predict whether the sale will happen or not. Model not trained properly.Data pattern not clear. That's why we got no purchase.
