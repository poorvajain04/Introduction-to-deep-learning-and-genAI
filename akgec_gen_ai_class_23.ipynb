```python
from google import genai
from google.colab import userdata

api_key = userdata.get("Gemini_api_key")
client = genai.Client(api_key=api_key)
```

```python
document=[
    "Flight cancellation is allowed before departure.",
    "Refund eligibility depends on cancellation time and fare type.",
    "Passengers need to have a bag of 15kg only."
]
```

```python
query="Can I cancel my flight and get refund?"
```

```python
prompt=f"""
Create 2 different search query for the question:
{query}
Return only the query, one per line.
"""
response=client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt,
)
gq=response.text.strip().split("\n")
print(gq)
```

```python
aq=[query]+gq
print(aq)
```

```python
def sd(q,documents):
  res=[]
  for d in documents:
    if ("flight" in q.lower() and "cancellation" in q.lower()):
      res.append(d)
  return res
```

```python
for q in aq:
  print("\nSearching",q)
  result=sd(q,document)
  for r in result:
    print("Retrieved:", r)
```

```python
def search_documents(query, documents):
    results = []
    for document in documents:
        if ("flight" in query.lower()) or \
           ("refund" in query.lower() and "refund" in document.lower()) or \
           ("cancel" in query.lower() and "cancellation" in document.lower()):
            results.append(document)
    return results
```

```python
for q in aq:
  print("\nSearching",q)
  result=search_documents(q,document)
  for r in result:
    print("Retrieved:", r)
```
