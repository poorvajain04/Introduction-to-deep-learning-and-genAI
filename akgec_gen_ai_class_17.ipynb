```python
from google.colab import userdata
from google import genai
```

```python
api_key = userdata.get("Gemini_api_key")
```

```python
client = genai.Client(api_key=api_key)
```

```python
prompt = "write a python code to check whether a number is prime or not."
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)
```

# **One Short Prompting**

```python
api_key=userdata.get("Gemini_api_key")
client=genai.Client(api_key=api_key)
prompt="""
example"
write a python code to add two numbers.
code:
a=20
b=30
print(a+b)
Now write python code to multiply two numbers
"""
response=client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)
print(response.text)
```

# **Personal Prompting**

```python
prompt="""
You are a senior interviewer HR at a company.
Interview a python developer.
Ask:
-five technical questions
-two coding questions including the concept of DP and graphs of leetcode hard level
-one scenario based question
"""
response=client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)
print(response.text)
```
