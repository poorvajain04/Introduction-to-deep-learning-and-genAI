```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
reviews=[
    "product is very good",
    "quality is excellent",
    "I like this product",
    "very bad quality",
    "poor quality",
    "I hate this item"
]
```

```python
label=np.array([1,1,1,0,0,0])
tokenizer=Tokenizer(num_words=100)
tokenizer.fit_on_texts(reviews)
sequences=tokenizer.texts_to_sequences(reviews)
X=pad_sequences(sequences,maxlen=5)
```

```python
model=Sequential()
model.add(Embedding(input_dim=100,output_dim=8,input_length=5))
model.add(LSTM(units=16))
model.add(Dense(units=1,activation="sigmoid"))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X,label,epochs=100,verbose=0)
```

```python
test_review=["this product is excellent"]
test_seq=tokenizer.texts_to_sequences(test_review)
test_pad=pad_sequences(test_seq,maxlen=5)
prediction=model.predict(test_pad)
print("Prediction",prediction)
```

```python
if prediction[0][0]>0.5:
  print("sentiment: positive")
else:
  print("sentiment:negative")
```

LSTM - Time Series Forecasting

```python
from tensorflow.keras.layers import LSTM,Dense
from tensorflow.keras.models import Sequential
import numpy as np
sales=np.array([100,120,130,150,170,180,220,240,260])
X=[]
Y=[]
for i in range (len(sales)-3):
  X.append(sales[i:i+3])
  Y.append(sales[i+3])

```

```python
X=np.array(X)
y=np.array(Y)
X=X.reshape(X.shape[0],X.shape[1],1)
```

```python
model=Sequential()
model.add(LSTM(32,activation="relu",input_shape=(3,1)))
model.add(Dense(1))
```

```python
model.compile(
    optimizer="adam",
    loss="mse"
)
```

```python
model.fit(X,y,epochs=300,verbose=0)
```

```python
test_input=np.array([220,240,260])
test_input=test_input.reshape(1,3,1)
prediction=model.predict(test_input)
print("Next Day Sales Prediction",prediction[0][0])
```
