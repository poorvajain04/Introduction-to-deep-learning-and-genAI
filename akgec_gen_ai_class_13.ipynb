**Simple chatbot using if else (Rule based)**

```python
user=input()
if user=="hello":
  print("Bot: Hi! How are you")
elif user=="How are you":
  print("Bot: I am good")
elif user=="Bye":
  print("Bot: Goodbye")
else:
  print("Bot: I didn't understand")
```

```python
sentence=input("Enter a sentence: ")
words=sentence.split()
print("words in sentence: ")
for i,word in enumerate(words,start=1):
  print(f"Word {i}: ",word)
```

```python
sentence=input("Enter a sentence")
chars=list(sentence)
print("character token")
for i,word in enumerate(chars,start=1):
  print(f"Character {i}",{word})
```
