# **Prompt Template Builder**

---


You are an expert {role}.
Design a reusable prompt template builder that can generate different prompts based on use inputs like role, topic and language without changing the original template.



```python
from google import genai
from google.colab import userdata
api_key=userdata.get("Gemini_api_key")
client=genai.Client(api_key=api_key)
template="""
You are an expert {role}.
Explain the topic "{topic}".
Language : {language}
give three real word example.
"""
role=input("Enter role: ")
topic=input("Enter topic: ")
language=input("Enter language: ")
prompt=template.format(
    role= {role},
    topic={topic},
    language={language}
)
print("Generated prompt: ")
response=client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)
print(response.text)
```

# **Create a prompt template builder that accept multiple user parameter and generate interview questions**

```python
template="""
You are an interviewer.
Interviewee knows {language} as coding language.
Interviewee has {prof} proficiency in the language.
Interviewee has an experience of {age} years.
Interviewee is applying for the {role} role.
Generate exactly 5 interview questions.
The questions should:
1. Match the candidate's proficiency.
2. Match their years of experience.
3. Be relevant to the job role.
4. Progress from easy to difficult.
Only output the questions.
"""
prompt=template.format(
    language=input("Enter programming language: "),
    prof=input("Enter proficiency (Beginner/Intermediate/Advanced): "),
    age=input("Enter years of experience: "),
    role=input("Enter job role: ")
)
response=client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)
print(response.text)
```

# **Loop Based Prompt Generator**



------------------------------------------------------------------


Create a loop based prompt generator that generates personalised study plan for multiple students


```python
students=[
    "Anna",
    "Ben"
]
for student in students:
  prompt=f"""
  You are an experienced teacher.students
  Create a personalised study plan for {student}.students
  Duration: 30 days.
  Subject: English, Hindi, Maths, Science, Social Science, Python, Java, C, C++, JavaScript, German.
  Requirements:
- Divide the plan day-wise.
- Allocate approximately equal study time for all subjects.
- Include revision after every 7 days.
- Add one small quiz every week.
- Mention daily learning goals.
"""
  response=client.models.generate_content(
      model="gemini-2.5-flash",
      contents=prompt
  )
  print(f"\n{'='*60}")
  print(f"Study Plan for {student}")
  print('='*60)
  print(response.text)
```

```python
topic = "Python"

formats = [
    "Plain Text",
    "Bullet",
    "Numbered List",
    "Table",
    "JSON",
    "Markdown",
    "Code Output"
]

for format_name in formats:
    prompt = f"""
Explain the topic: {topic}.

Output Format:
{format_name}
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    print("-" * 60)
    print(f"Format: {format_name}")
    print("-" * 60)
    print(response.text)
    print()
```
